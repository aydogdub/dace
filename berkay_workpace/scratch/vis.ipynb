```python
import dace
import cupy as cp

from IPython.display import Code
from dace.config import Config
```

```python
@dace.program
def cpuRed(A: dace.uint32[32], B:dace.uint32[32]):
    for i in dace.map[0:32]:
        B[i] = sum(A)
```

```python
sdfg = cpuRed.to_sdfg()
sdfg.compile()
sdfg
```

```python
Code(sdfg.generate_code()[0].clean_code, language='cpp')
```
